```python
credits.shape
```

```python
# Merging dataFrame movies and credits into a single dataframe movies
movies = movies.merge(credits,on= 'title')
```

```python
print(movies.shape)
movies.head()
```

```python
# genres
# id
# keywords
# overview
# title
# cast
# crew
# Keeping only relevant columns in the dataset
movies = movies[['movie_id','genres','keywords','overview','title','cast','crew']]
movies.head()
```

```python
# Since null rows are negligible so we are dropping it
print(movies.isnull().sum())
movies.dropna(inplace=True)
print(movies.isnull().sum())
```

```python
# helper function for converting string of genres,keywords,crew,cast into list of all these.
import ast
def convert(obj):
  lst = []
  for i in ast.literal_eval(obj):
    lst.append(i['name'])
  return lst
```

```python
ast.literal_eval(movies.cast[0])[:3]
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.keywords = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
  lst = []
  for i in ast.literal_eval(obj)[:3]:
    lst.append(i['name'])
  return lst
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies['crew'][0]
```

```python
def convertcrew(obj):
  lst = []
  for i in ast.literal_eval(obj):
    if i["job"] == "Director":
      lst.append(i['name'])
      break
  return lst
```

```python
movies['crew'] = movies['crew'].apply(convertcrew)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
#  removing space from each item in list from the following columns
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#  Creation of a tag column which will be the concatination of columns genres,overview,keywords,cast,crew
movies['tags'] = movies['overview'] + movies['genres'] + movies['cast'] + movies['crew'] + movies['keywords']
```

```python
movies.head()
```

```python
# Creating a new dataframe
new_df = movies[['movie_id','title','tags']]
new_df
```

```python
# Converting the list to string
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df
```

```python
# converting to lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english',max_features=5000)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movie_list:
    print(new_df.iloc[i[0]].title)
```

```python
type(new_df['title'].values)
```

```python
recommend('Spider-Man')
```
